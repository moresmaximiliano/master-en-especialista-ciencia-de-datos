```python
# eliminar filas
df = df.drop([26,27]) # pongo los índices
print(df)
print("-------------------------------------------------------------")
# filtrar filas de una columna (por condiciones numéricas)
df = df[df['PGB'] >= 0.0] # el nuevo df solo contiene las filas en donde el PGB sea mayor o igual a 0
print(df)
print("-------------------------------------------------------------")
# filtrar filas de una columna (por cadenas)
df2 = df[df['Departamentos'].str.contains('Ca')] # me filtra en un nuevo df los departamentos que contengan "ca"
print(df2)
print("-------------------------------------------------------------")
# agregar filas al df
new_row = pd.DataFrame({'Consumo': [12404], 'PGB': [177644],'Departamentos': ['Tulumba']}) # creamos las filas a agregar
new_row2 = pd.DataFrame({'Consumo': [101052], 'PGB': [1491739],'Departamentos': ['Unión']})
df = pd.concat([df, new_row, new_row2], ignore_index=True)
print(df)
```
